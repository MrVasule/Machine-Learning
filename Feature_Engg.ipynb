```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
```

```python
df
```

```python
x = df.data
y= df.target
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
x_reduced = pca.fit(x).transform(x)
```

```python
x_reduced
```

```python
x_reduced = pd.DataFrame(x_reduced,columns=['x1','x2'])
```

```python
x_reduced
```

```python
sns.scatterplot(x='x1',y='x2',data=x_reduced,hue=y)
plt.show()
```

# Feature Selection
# a - Low Varience 

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.6)
```

```python
x_vt = sel.fit_transform(x)
```

```python
x_vt
```

```python
x_vt = pd.DataFrame(x_vt,columns=['x1','x2'])
```

```python
x_vt
```

```python
sns.scatterplot(x='x1',y='x2',data=x_vt,hue=y)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest,SelectPercentile,chi2
```

```python
x_sb = SelectKBest(chi2,k=2).fit_transform(x,y)
```

```python
x_sb
```

```python
x_sb = pd.DataFrame(x_sb,columns=['x1','x2'])
```

```python
x_sb
```

```python
sns.scatterplot(x='x1',y='x2',data=x_sb,hue=y)
plt.show()
```

```python
x_sp = SelectPercentile(chi2,percentile=50).fit_transform(x,y)
```

```python
x_sp
```

#  Select From model

```python
from sklearn.svm import LinearSVC
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
clf = LinearSVC()
```

```python
clf.fit(x,y)
```

```python
model = SelectFromModel(clf,prefit=True)
```

```python
x_sfm = model.transform(x)
```

```python
x_sfm
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x_sfm[:,0],x_sfm[:,1],x_sfm[:,2],c=y,cmap='rainbow')
plt.show()
```
