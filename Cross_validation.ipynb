```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
datasets = load_iris()
```

```python
datasets
```

```python
iris = pd.DataFrame(data=datasets.data,columns=datasets.feature_names)
```

```python
iris['label'] = datasets.target
```

```python
iris
```

```python
X = np.array(iris.drop(['label'],axis=1))
```

```python
y = np.array(iris['label'])
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracy = cross_val_score(model,X,y,scoring='accuracy',cv=10)
```

```python
accuracy
```

```python
accuracy.mean()
```

#second method

```python
from sklearn.model_selection import KFold
```

```python
cv = KFold(n_splits=10,shuffle=True)
```

```python
score = []
for train_index,test_index in cv.split(X):
    print('Train_Index: ',train_index)
    print('Test Index: ',test_index)
    X_train,X_test,y_train,y_test = X[train_index],X[test_index],y[train_index],y[test_index]
    model.fit(X_train,y_train)
    score.append(model.score(X_test,y_test))
```

```python
score
```

```python
np.mean(score)
```
