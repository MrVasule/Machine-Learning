```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_style('whitegrid')
```

```python
ar = pd.read_csv('countrycluster.csv')
```

```python
#Dendrogrom
```

```python
x = ar[['Latitude','Longitude']]
```

```python
sns.clustermap(x)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(distance_threshold=20,n_clusters=None)
```

```python
model.fit(x)
```

```python
id_cl = model.fit_predict(x)
```

```python
id_cl
```

```python
plt.scatter(ar['Longitude'],ar['Latitude'],c=id_cl,cmap='rainbow')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
