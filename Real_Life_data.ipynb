```python
data_no_multicolinearity = data_cleaned.drop(['Year'],axis=1)
```

```python
data_no_multicolinearity
```

```python

# deal Categorical Variables

```

```python
data_no_multicolinearity.describe(include='all')
```

```python
data_no_multicolinearity
```

```python
data_with_dummies = pd.get_dummies(data_no_multicolinearity,drop_first=True)
```

```python
data_with_dummies
```

```python
targets = data_with_dummies['log_price']  #y (dependent variables)
```

```python

inputs = data_with_dummies.drop(['log_price'],axis=1) #x1,x2,x3,......,x17 (independent)
```

```python
inputs
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
StandardScaler()
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
inputs_scaled
```

```python
a = pd.DataFrame(inputs_scaled[3:6])
```

```python
a
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(inputs_scaled,targets,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
LinearRegression()
```

```python
reg.score(X_train,y_train) #training score
```

```python
reg.score(X_test,y_test) #testing score
```

```python
reg.coef_ #b1,b2,b3,....
```

```python
reg.intercept_ #b0
```

```python
y_hat = reg.predict(X_test)
```

```python
y_hat
```

```python
y_test
```

```python
sns.distplot(y_test-y_hat)
plt.show()
```

```python
plt.scatter(y_test,y_hat)
plt.xlabel('Actual data')
plt.ylabel('Predicted data')
plt.show()
```

```python

evaluation = pd.DataFrame(np.array(y_test).reshape(-1,1),columns=['Actual Data'])
```

```python
evaluation['predicted data'] = y_hat
```

```python
evaluation
```

```python
evaluation['Non Log Prediction'] = np.exp(y_hat)
```
