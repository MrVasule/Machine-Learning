```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import  load_boston
```

```python
df = load_boston()
```

```python
df
```

```python
dataset = pd.DataFrame(data=df.data,columns=df.feature_names)
```

```python
dataset['price'] = df.target
```

```python
dataset
```

```python
x = dataset.iloc[:,0:-1].values
```

```python
x
```

```python
y = dataset.iloc[0:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lin_regressor = LinearRegression()
```

```python
mse = cross_val_score(lin_regressor,x,y,scoring='neg_mean_squared_error',cv=5)
```

```python
mse
```

```python
np.mean(mse)
```

# Lasso

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()
```

```python
hyperparameter = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,1,2,4,5,10,20,30,40,50,60,70,55,100]}
```

```python
lasso_regressor = GridSearchCV(lasso,hyperparameter,scoring='neg_mean_squared_error',cv=5)
```

```python
lasso_regressor
```

```python
lasso_regressor.fit(x,y)
```

```python
lasso_regressor.best_score_
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
```

```python
ridge_regressor = GridSearchCV(ridge,hyperparameter,scoring='neg_mean_squared_error',cv=5)
```

```python
ridge_regressor.fit(x,y)
```

```python
ridge_regressor.best_score_
```
