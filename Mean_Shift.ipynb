```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_blobs
```

```python
centers = [[2,2,2],[7,7,7],[5,13,13]]
```

```python
x,_ = make_blobs(n_samples=150,centers=centers)
```

```python
x
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2])
plt.show()
```

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift()
```

```python
ms.fit(x)
```

```python
identified_cluster = ms.fit_predict(x)
```

```python
identified_cluster
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],c=identified_cluster,cmap='rainbow')
plt.show()
```

#  KMeans

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=6)
```

```python
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
```

```python
identified_clusters
```

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],c=identified_clusters,cmap='rainbow')
plt.show()
```

# ELBOW Method

```python
wcss = []
for i in range(1,7):
  kmeans = KMeans(i)
  kmeans.fit(x)
  wcss_inter = kmeans.inertia_
  wcss.append(wcss_inter)
```

```python
wcss
```

```python
plt.plot(range(1,7),wcss)
plt.title('ELBOW Method')
plt.xlabel('Num of Cluster')
plt.ylabel('WCSS')
plt.show()
```
