```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
sns.set_style('whitegrid')
from sklearn.datasets import make_blobs
```

```python
centers = [[0.5,2],[-1,-1],[1.5,-1]]
```

```python
x, _ = make_blobs(n_samples=400, centers=centers,cluster_std=0.5)
```

```python
x
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.4,min_samples=10)
```

```python
db.fit(x)
```

```python
identified_cluster = db.fit_predict(x)
```

```python
identified_cluster
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=identified_cluster,cmap='rainbow')
plt.show()
```
