```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,1].values
```

```python
y = dataset.iloc[:,2].values
```

```python
plt.scatter(x,y)
plt.xlabel('Level')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor()
```

```python
reg.fit(x.reshape(-1,1),y)
```

```python
reg.predict(np.array([7.5]).reshape(-1,1))
```

```python
reg.predict(np.array([7.4]).reshape(-1,1))
```

```python
reg.predict(np.array([5.6]).reshape(-1,1))
```

```python
x_grid = np.arange(min(x),max(x),0.01)
```

```python
x_grid
```

```python
x_grid.shape
```

```python
x_grid = x_grid.reshape(900,1)
```

```python
x_grid
```

```python

plt.scatter(x,y, color='red')
plt.plot(x_grid,reg.predict(x_grid))
plt.title('DecisionTree')
plt.xlabel('level')
plt.ylabel('salary')
plt.show()
```
